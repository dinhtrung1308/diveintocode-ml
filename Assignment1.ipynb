**[Problem 1] Create using the exponentiation arithmetic operator**

```

THICKNESS = 0.00008
folded_thickness = THICKNESS*2**43
print("Thickness: {} meters".format(folded_thickness))
```

**[Problem 2] Unit Conversion**

```
print("Thickness: {: .2f} * 10,000 kilometers".format(folded_thickness/(1000*10000) ))


```

**[Problem 3] Create using a for statement**

```
times=43
folded_thickness=THICKNESS
for i in range(0,times):
  folded_thickness=folded_thickness*2
  
print("Thickness: {} meters".format(folded_thickness))
```

**[Problem 4] Comparison of calculation time**


```
import time
start = time.time()
THICKNESS = 0.00008
folded_thickness = THICKNESS*2**43
elapsed_time = time.time() - start
print("time : {}[s] using exponentiation arithmetic operator".format(elapsed_time))

start = time.time()
THICKNESS = 0.00008
times=43
folded_thickness=THICKNESS
for i in range(0,times):
  folded_thickness=folded_thickness*2
elapsed_time = time.time() - start
print("time : {}[s] using for statement".format(elapsed_time))
 
```

**[Problem 5] Saving to a list**


```
thicknessList = []

for i in range(0,44):
    thicknessList.append(THICKNESS*(2**i))
print(thicknessList)
print(len(thicknessList))
```

**Problem 6: Displaying a line graph**


```
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thicknessList) # Enter the variable name of the list in "List name"
plt.show()
```

**[Problem 7] Customizing graphs**

```
"""
Display a orange line graph.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thicknessList, color='orange')
plt.show()
```

```
"""
Display a orange dashed green graph.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thicknessList, color='green', linestyle='dashed')
plt.show()
```

```
"""
Display a dotted navy graph.
And change the size.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds",size= 30)
plt.ylabel("thickness[m]", size=30)
plt.tick_params(labelsize=15) 
plt.plot(thicknessList, color='navy', linestyle='dotted')
plt.show()
```
